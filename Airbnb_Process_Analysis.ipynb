```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_excel('/content/1730285881-Airbnb_Open_Data.xlsx')
```

```python
df.drop_duplicates(inplace = True)
```

```python
# Drop columns with insufficient data (as identified in the notebook)
df.drop(['house_rules', 'license'], axis = 1, inplace=True)
```

```python
# Remove dollar signs and commas from price and service fee columns
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False)
df['service fee'] = df['service fee'].astype(str).str.replace('$', '', regex=False)
```

```python
df['price'] = df['price'].str.replace(',', '', regex=False)
df['service fee'] = df['service fee'].str.replace(',', '', regex=False)
```

```python
# Rename the price and service fee columns
df.rename(columns = {
    'price' : 'price_$',
    'service fee' : 'service_fee_$'
}, inplace = True)
```

```python
# Drop all records with any remaining missing values
df.dropna(inplace = True)
```

```python
# Change mismatched data types
df['price_$'] = df['price_$'].astype(float)
df['service_fee_$'] = df['service_fee_$'].astype(float)
df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'])
df['Construction year'] = df['Construction year'].astype(int)
```

```python
# Correct the spelling of 'brookln' to 'Brooklyn'
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'
```

```python
df.drop(df[df['availability 365'] > 500].index, inplace=True)
```

```python
# --- Q1: Property Types and Count (Image 1000034255.jpg) ---
print("\n--- Property Types Count ---")
property_types = df['room type'].value_counts().to_frame().reset_index()
property_types.columns = ['Room Type', 'Count']
print(property_types)

plt.figure(figsize=[8, 6])
room_type_bar = plt.bar(property_types['Room Type'], property_types['Count'])
plt.bar_label(room_type_bar, labels = property_types['Count'], padding = 4)
plt.ylim(0, property_types['Count'].max() * 1.1)
plt.xlabel("Room Type")
plt.ylabel("Room Type Count")
plt.title("Property Types and their count in the Dataset")
plt.savefig('room_type_count_bar.png')
plt.close()
print("Plot saved: room_type_count_bar.png")
```

```python
# --- Q2: Highest number of listings by Neighborhood Group (Image 1000034256.jpg) ---

hood_group = df['neighbourhood group'].value_counts().to_frame()
print(hood_group)
```

```python
hood_group_bar = plt.bar(hood_group.index, hood_group['count'])
plt.bar_label(hood_group_bar, labels = hood_group['count'], padding = 4)
plt.ylim([0, 40000])
plt.xlabel('Neighborhood Group')
plt.ylabel('Number of Listings')
plt.xticks(rotation = 45)
plt.title('Which Neighborhood Groups has the highest number of Listings')
plt.tight_layout()
plt.savefig('neighbourhood_group_bar.png')
plt.close()
print("Plot saved: neighbourhood_group_bar.png")
```

```python
print("\n--- Average Price by Neighborhood Group ---")
avg_price = df.groupby(['neighbourhood group'])['price_$'].mean().sort_values(ascending = False).to_frame().reset_index()
avg_price.columns = ['neighbourhood group', 'price_$']
print(avg_price)

plt.figure(figsize=[10, 6])
avg_price_bar = plt.bar(avg_price['neighbourhood group'], avg_price['price_$'])
plt.bar_label(avg_price_bar, labels = round(avg_price['price_$'], 2), padding = 4)
plt.ylim([0, avg_price['price_$'].max() * 1.1])
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price per Listings ($)')
plt.xticks(rotation = 45)
plt.title('Average Price per Listings ($) in each Neighborhood Group')
plt.tight_layout()
plt.savefig('avg_price_by_neighborhood_bar.png')
plt.close()
print("Plot saved: avg_price_by_neighborhood_bar.png")
```

```python
print("\n--- Average Price by Construction Year (Line Plot) ---")
avg_price_by_year = df.groupby(df['Construction year'])['price_$'].mean().to_frame()

plt.figure(figsize=[10, 6])
avg_price_by_year.plot(ax=plt.gca(), legend=False)
plt.xlabel('Construction Year')
plt.ylabel('Average Price ($)')
plt.title('Average Price ($) for Properties in each Construction Year')
plt.grid(True, axis='y', linestyle='--')
plt.tight_layout()
plt.savefig('avg_price_by_construction_line.png')
plt.close()
print("Plot saved: avg_price_by_construction_line.png")
```

```python
# --- Q5: Top 10 hosts by calculated host listing count (Image 1000034260.jpg) ---
print("\n--- Top 10 Hosts by Calculated Host Listing Count ---")
hosts = df.groupby(['host name'])['calculated host listings count'].sum().sort_values(ascending = False).nlargest(10).to_frame().reset_index()
hosts.columns = ['Host Name', 'calculated host listings count']
print(hosts)

plt.figure(figsize=[12, 7])
hosts_bar = plt.bar(hosts['Host Name'], hosts['calculated host listings count'])
plt.bar_label(hosts_bar, labels = hosts['calculated host listings count'], label_type = 'edge', padding = 3)
plt.xlabel('Hosts Name')
plt.ylabel('Calculated Host listings count')
plt.xticks(rotation = 80)
plt.ylim([0, hosts['calculated host listings count'].max() * 1.1])
plt.title('Top 10 Hosts by Calculated Host Listings Count')
plt.tight_layout()
plt.savefig('top_10_hosts_bar.png')
plt.close()
print("Plot saved: top_10_hosts_bar.png")
```

```python
print("\n--- Average Review Rate by Host Verification Status ---")
review = df.groupby(['host_identity_verified'])['review rate number'].mean().sort_values(ascending = False).to_frame().reset_index()
review.columns = ['host_identity_verified', 'review rate number']
print(review)
```

```python
print("\n--- Correlation: Price vs. Service Fee ---")
correlation_price_service = df['price_$'].corr(df['service_fee_$'])
print(f"Pearson correlation coefficient (price vs. service fee): {correlation_price_service}")
```

```python
sns.regplot(data=df, x = 'price_$', y = 'service_fee_$')
plt.xlabel('Price ($)')
plt.ylabel('Service Fee ($)')
plt.title('A Regression Plot showing the Correlation of the Price of a Listing and its Service Fee')
plt.tight_layout()
plt.savefig('price_vs_service_fee_regplot.png')
plt.close()
print("Plot saved: price_vs_service_fee_regplot.png")
```

```python
print("\n--- Average Review Rate by Neighborhood Group and Room Type ---")
ARRN = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().to_frame()
print(ARRN)
```

```python

sns.barplot(data = df, x = 'neighbourhood group', y = 'review rate number', hue = 'room type', errorbar=None)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Review Rate')
plt.title('Average Review Rate for each Room/Property Type in each Neighbourhood Group')
plt.legend(title='room type')
plt.tight_layout()
plt.savefig('avg_review_rate_by_hood_room_bar.png')
plt.close()
print("Plot saved: avg_review_rate_by_hood_room_bar.png")
```

```python
print("\n--- Correlation: Calculated Host Listings Count vs. Availability 365 ---")
correlation_listings_availability = df['calculated host listings count'].corr(df['availability 365'])
print(f"Pearson correlation coefficient (listings vs. availability): {correlation_listings_availability}")
```

```python
sns.regplot(data=df, x = 'calculated host listings count', y = 'availability 365')
plt.xlabel('Calculated Host listings')
plt.ylabel('Availability 365')
plt.title('A Regression Plot of the Relationship between Calculated Host Listings Count and Availability 365')
plt.tight_layout()
plt.savefig('listings_vs_availability_regplot.png')
plt.close()
print("Plot saved: listings_vs_availability_regplot.png")
```
